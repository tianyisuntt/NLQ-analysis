```python
from bs4 import BeautifulSoup
def cleanText(text):
    text = BeautifulSoup(text, "lxml").text
    text = re.sub(r'\|\|\|', r' ', text) 
    text = re.sub(r'http\S+', r'<URL>', text)
    text = text.lower()
    text = text.replace('x', '')
    text = text.replace('\n', '')
    text=text.replace("*\t","")
    return text
data_long['text_long'] = data_long['text_long'].apply(cleanText)
```

```python
import torch.nn as nn
```

```python
# The linear classifier given by  CoLA-baselines model
class LinearClassifier(nn.Module):
    """
    A basic linear classifier for acceptability judgments.
    Input sentence embedding (with size = 2*encoding_size,
    factor of 2 comes from bidirectional LSTM encoding)
    Hidden layer (encoding 2 * encoding_size * hidden_size)
    Output layer (hidden_size * 1)
    """
    def __init__(self, hidden_size, encoding_size, dropout=0.001):
        super(LinearClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.dropout = nn.Dropout(dropout)
        self.enc2h = nn.Linear(encoding_size, self.hidden_size)
        self.h20 = nn.Linear(self.hidden_size, 1)
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()
        self.softmax = nn.Softmax()
    
    def forward(self, sentence_vecs):
        hidden = self.tanh(self.dropout(self.enc2h(sentence_vecs)))
    
        
        out = self.sigmoid(self.h20((hidden)))
        return out
```

```python
# train the model 
Classifier=LinearClassifier(100,600)
Classifier.double()
optimizer=torch.optim.Adam(Classifier.parameters(),lr=0.0001)
batch_size=50
epochs=300
for i in range(epochs):
          Classifier.train()
          train_loss=0.0
          for k in range(int(len(X_train)/batch_size)):
                        if (k+1)*batch_size>len(X_train):
                                    train_set=X_train[k*batch_size:]
                                    label_set=y_train[k*batch_size:]
                        else:
                                    train_set=X_train[k*batch_size:(k+1)*batch_size]
                                    label_set=y_train[k*batch_size:(k+1)*batch_size]
                      
                        train_set=torch.FloatTensor([i for i in train_set]).double()
                  
                        label_set=torch.FloatTensor([i for i in label_set]).double()
                      
                        output=Classifier(train_set)
                        loss=loss_binary(output,label_set.view(50,1))
                        optimizer.zero_grad=0
                        loss.backward()
                        #torch.nn.utils.clip_grad_norm_(Classifier.parameters())
                        optimizer.step()
                        train_loss+=loss.item()
          average_train_loss=train_loss/len(X_train)
          print("epochs:{} train_loss:{}".format(i+1,average_train_loss))
          Classifier.eval()   
          test_loss=0.0
          for k in range(int(len(X_test))):
                        test_set=X_test[k:k+1]
                        test_label_set=y_test[k:k+1]
                        test_set=torch.FloatTensor([i for i in test_set]).double()
                  
                        test_label_set=torch.FloatTensor([i for i in test_label_set]).double()
                        output=Classifier(test_set)
                        loss=loss_binary(output,test_label_set.view(1,1))    
                        test_loss+=loss.item()
          average_test_loss=test_loss/len(X_test)
          print("test_loss:{}".format(average_test_loss))
                        
```

```python
import numpy as np
def similarity_scores_cosin(sentence_one, sentence_two):
                       sentence_one=torch.tensor(sentence_one)
                       sentence_two=torch.tensor(sentence_two)
                       dot_product=torch.dot(sentence_one,sentence_two)
                       L_one=torch.sqrt(torch.sum(torch.square(sentence_one)))
                       L_two=torch.sqrt(torch.sum(torch.square(sentence_two)))
                       return dot_product/(L_one*L_two)
def manhattan_distances(sentence_one, sentence_two):
                       dit=sentence_one-sentence_two
                       return sum(abs(dit))
def euclidean_distances(sentence_one,sentence_two):
                       dit=sentence_one-sentence_two
                       dit_squares=np.square(dit)
                       return(np.sqrt(np.sum(dit_squares)))
```

```python
def pearson_scores(X,Y):
            return np.cov(X, Y)/(np.std(X)*np.std(Y))

def spearman_scores(X,Y):
             rank_x=[sorted(X).index(i)+1 for i in X]
             
             rank_y=[sorted(Y).index(i)+1 for i in Y]
             d=rank_x-rank_y
             sum_sqrt=d^2
             p=1-6*sum(sum_sqrt)/(len(X)*(len(X)^2-1))
             return p
```

```python
import torch
from scipy import stats
```

```python
def EmbeddingSimilarityEvaluator(sentences_one, sentences_two,labels):
        dic={}
        cosine_scores=[similarity_scores_cosin(sentences_one[i],sentences_two[i]) for i,_ in enumerate(sentences_one)]
        manhattan_distance=[manhattan_distances(sentences_one[i],sentences_two[i]) for i,_ in enumerate(sentences_one)]
        euclidean_distance=[euclidean_distances(sentences_one[i],sentences_two[i]) for i,_ in enumerate(sentences_one)]
         
      
        eval_pearson_cosine =stats.pearsonr(labels, cosine_scores)
       
        eval_spearman_cosine=stats.spearmanr(labels, cosine_scores)

        eval_pearson_manhattan=stats.pearsonr(labels, manhattan_distance)
        eval_spearman_manhattan=stats.spearmanr(labels, manhattan_distance)

        eval_pearson_euclidean=stats.pearsonr(labels, euclidean_distance)
        eval_spearman_euclidean=stats.spearmanr(labels, euclidean_distance)

        dic["eval_pearson_cosine"]=eval_pearson_cosine
        dic["eval_spearman_cosine"]=eval_spearman_cosine
        dic["eval_pearson_manhattan"]=eval_pearson_manhattan
        dic["eval_spearman_manhattan"]=eval_spearman_manhattan
        dic["eval_pearson_euclidean"]=eval_pearson_euclidean
        dic["eval_spearman_euclidean"]=eval_spearman_euclidean
        return dic
```

```python
#Removing the html strips
def strip_html(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

#Removing the square brackets
def remove_between_square_brackets(text):
    return re.sub('\[[^]]*\]', '', text)

#Removing the noisy text
def denoise_text(text):
    text = strip_html(text)
    text = remove_between_square_brackets(text)
    return text
#Apply function on review column
train_texts=[denoise_text(i[0]) for i in train_texts]
```

```python
def remove_special_characters(text, remove_digits=True):
    pattern=r'[^a-zA-z0-9\s]'
    text=re.sub(pattern,'',text)
    return text
def simple_stemmer(text):
    ps=nltk.porter.PorterStemmer()
    text= ' '.join([ps.stem(word) for word in text.split()])
    return text
train_texts=[remove_special_characters(i) for i in train_texts]
train_texts[0]
```

```python
def remove_stopwords(text, is_lower_case=False):
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    if is_lower_case:
        filtered_tokens = [token for token in tokens if token not in stopword_list]
    else:
        filtered_tokens = [token for token in tokens if token.lower() not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text
```

```python
import nltk
nltk.download('punkt')

def tokenize_text(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            if len(word) < 2:
                continue
            tokens.append(word.lower())
    return tokens
train_texts=[tokenize_text(i) for i in train_texts]
```

```python
class LinearClassifier(nn.Module):
    """
    A basic linear classifier for acceptability judgments.
    Input sentence embedding (with size = 2*encoding_size,
    factor of 2 comes from bidirectional LSTM encoding)
    Hidden layer (encoding 2 * encoding_size * hidden_size)
    Output layer (hidden_size * 1)
    """
    def __init__(self, hidden_size, encoding_size, dropout=0.001):
        super(LinearClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.dropout = nn.Dropout(dropout)
        self.enc2h = nn.Linear(encoding_size, self.hidden_size)
        self.h20 = nn.Linear(self.hidden_size, 1)
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()
        self.softmax = nn.Softmax()
    
    def forward(self, sentence_vecs):
        hidden = self.tanh(self.dropout(self.enc2h(sentence_vecs)))
    
        
        out = self.sigmoid(self.h20((hidden)))
        return out
```

```python
from torch import nn
import torch
loss_binary=torch.nn.BCELoss()
```

```python
Classifier=LinearClassifier(50,800)
Classifier.double()
optimizer=torch.optim.Adam(Classifier.parameters(),lr=0.00000001)
batch_size=50
epochs=300
for i in range(epochs):
          Classifier.train()
          train_loss=0.0
          for k in range(int(len(X_train)/batch_size)):
                        if (k+1)*batch_size>len(X_train):
                                    train_set=X_train[k*batch_size:]
                                    label_set=y_train[k*batch_size:]
                        else:
                                    train_set=X_train[k*batch_size:(k+1)*batch_size]
                                    label_set=y_train[k*batch_size:(k+1)*batch_size]
                      
                        train_set=torch.FloatTensor([i for i in train_set]).double()
                  
                        label_set=torch.FloatTensor([i for i in label_set]).double()
                      
                        output=Classifier(train_set)
                        loss=loss_binary(output,label_set.view(50,1))
                        optimizer.zero_grad=0
                        loss.backward()
                        #torch.nn.utils.clip_grad_norm_(Classifier.parameters())
                        optimizer.step()
                        train_loss+=loss.item()
          average_train_loss=train_loss/len(X_train)
          print("epochs:{} train_loss:{}".format(i+1,average_train_loss))
          Classifier.eval()   
          test_loss=0.0
          for k in range(int(len(X_test))):
                        test_set=X_test[k:k+1]
                        test_label_set=y_test[k:k+1]
                        test_set=torch.FloatTensor([i for i in test_set]).double()
                  
                        test_label_set=torch.FloatTensor([i for i in test_label_set]).double()
                        output=Classifier(test_set)
                        loss=loss_binary(output,test_label_set.view(1,1))    
                        test_loss+=loss.item()
          average_test_loss=test_loss/len(X_test)
          print("test_loss:{}".format(average_test_loss))
```
