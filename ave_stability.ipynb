```python
category_K = pd.read_csv('/kaggle/input/batch6allscores/batch6_category_Krippendroffis_alpha.csv')
category_F = pd.read_csv('/kaggle/input/batch6allscores/batch6_category_Fleiss_kappa.csv')
category_F.rename(columns={'Unnamed: 0': 'category'}, inplace=True)
category_stability = pd.read_csv('/kaggle/input/batch6allscores/batch6_category_stability.csv')

type_stability = pd.read_csv('/kaggle/input/batch6allscores/batch6_type_stability.csv')
```

```python
category_result = pd.merge(category_F, category_K, on = 'category')
#category_result.to_csv('batch6_category_result.csv', index = True)
category_result
```

```python
batch1_stability = category_stability
batch1_stability.rename(columns={'Unnamed: 0': 'error_category'}, inplace=True)
batch1_stability = batch1_stability.set_index('error_category')
interdivunion2=[]
count0=0
#batch1_stability.iloc[4,:]
#batch1_stability.iloc[5,:]
for i in range(0,len(batch1_stability.iloc[5,:])):
    if float(batch1_stability.iloc[5,:][i]) ==0:
        count0 +=1
    if float(batch1_stability.iloc[5,:][i]) !=0:
        interdivunion2.append(float(batch1_stability.iloc[4,:][i])/float(batch1_stability.iloc[5,:][i]))
ave = sum(interdivunion2)/(len(interdivunion2)-count0)
print('batch1 average stability, where coder intersections are 2, is:', ave)
```

```python
category_F.rename(columns={'Unnamed: 0': 'category'}, inplace=True)
category = category_K.merge(category_F, on ='category')
stability = category_stability[6:].T
category = category.merge(stability, left_on ='category', right_on = stability.index, how = 'left')
category = category.rename(columns={6:'stability'})
category.stability.iloc[len(category)-1:] = ave
category = category[['category','Fleiss_kappa','Krippendroffis_alpha','stability']]
category.to_csv('batch6_category_results.csv')
category
```
